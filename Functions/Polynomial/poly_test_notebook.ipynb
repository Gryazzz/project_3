```python
from polynomial import getPoly
from polynomial import plotPoly
import matplotlib.pyplot as plt
import pandas as pd

path = 'resources/bus.csv'
bus_df = pd.read_csv(path)

bus_df['Date'] = pd.to_datetime(bus_df['Date'],format='%Y%m')

p = getPoly(bus_df['Date'].values, bus_df['Ridership'].values, 3)
f = plotPoly(bus_df['Date'].values, bus_df['Ridership'].values,p,show=True)
```

```python
f
ax1 = f.add_axes([0,0,1,1])
ax1.set_ylabel('Ridership (million)',weight='bold',size='20')

plt.show()
```

```python
path = 'resources/combined.csv'
combined_df = pd.read_csv(path)

combined_df['Date'] = pd.to_datetime(combined_df['Date'])
combined_df.head()
```
