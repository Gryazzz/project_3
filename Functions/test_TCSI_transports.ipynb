```python
df = pd.DataFrame(pd.read_csv('../results/combined.csv'))
#df2 = pd.DataFrame(pd.read_csv('../results/metro_kpi.csv'))
#df = pd.merge(df1, df2, on='Date', how='outer')
#df['ElevatorAvail'] = df['ElevatorAvail'].str.rstrip('%').astype('float')
#df['ElevatorAvail'] = df['ElevatorAvail'] /100
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
features = [column for column in df]
years = [2013, 2011, 2015, 2016]
features
```

```python
x = 0
f, ax = plt.subplots(figsize=(10,5))
for feature in features:
    if feature != 'Date':
        df1 = df[['Date', feature]]
        df1 = df1.dropna(how='any')
        time = df1.iloc[:,0].tolist()
        feat = df1.iloc[:,1].tolist()
        #pprint(time)
        m, tcsidf = TCSI(time, feat, years[x])
        #print(years[x])
        #print(tcsidf)
        tcsidf.plot(x='Date', y='trend', ax=ax, linewidth=2)
        #ax = tcsidf.plot(x='Date', y='trend')
        plt.title(f'Trend in {feature}')
        x +=1
```
