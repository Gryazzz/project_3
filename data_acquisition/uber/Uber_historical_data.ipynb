```python
months = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct',
          11: 'Nov', 12: 'Dec'}
```

```python
dist_data = {}
geodata = pd.read_json('raw_data/washington_DC_censustracts.json')
for item in geodata['features']:
    mid = int(item['properties']['MOVEMENT_ID'])
    #addr = item['properties']['DISPLAY_NAME'].split(',')
    address = item['properties']['DISPLAY_NAME']
    
    #try:
        #address = [addr[0], addr[1], addr[2], addr[3]]
    #except(IndexError):
        #try:
            #address = [addr[0], addr[1], addr[2], 0]
        #except(IndexError):
            #address = [addr[0], addr[1], 0, 0]
    dist_data[mid] = address
    #print(addr)
```

```python
geodata['features'][1]['geometry']['coordinates'][0][0][-1]
```

```python
dist_data_df = pd.DataFrame([dist_data]).transpose()
dist_data_df.head()
```

```python
df = m_2016_1_df
for ind, row in m_2016_1_df.iterrows():
    
    df.at[ind, 'Month'] = months[row['month']]
    if row['sourceid'] in dist_data:
        df.at[ind, 'Departure'] = dist_data[row['sourceid']]
    if row['dstid'] in dist_data:
        df.at[ind, 'Destination'] = dist_data[row['dstid']]
    
```
